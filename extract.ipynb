## video_to frame_png


```python
import cv2
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
def video_frames(input_path, output_path):
    # extract frames from a video and save to directory as 'x.png' where 
    # x is the frame indexZ
    vidcapture = cv2.VideoCapture(input_path)
    c= 0
    while vidcapture.isOpened():
        success, image = vidcapture.read()
        if success:
            cv2.imwrite(os.path.join(output_path, '%d.png') % c, image)
            c += 1
        else:
            break
    cv2.destroyAllWindows()
    vidcapture.release()
#    video_frames('C:/Users/mEtrOpOliS/Desktop/videos/1.avi', 'C:/Users/mEtrOpOliS/Desktop/imgs12'
```

## video_frame-jpg 

```python
def video_frame(input_path, output_path):
    # extract frames from a video and save to directory as 'x.jpg' where 
    # x is the frame indexZ
    vidcap = cv2.VideoCapture(input_path)
    c= 0
    while vidcap.isOpened():
        success, image = vidcap.read()
        if success:
            vidcap.set(cv2.CAP_PROP_POS_MSEC,c*1000)
            cv2.imwrite(os.path.join(output_path, '%d.JPG') % c, image)
            c += 1
            
        else:
            break
    cv2.destroyAllWindows()
    vidcap.release()

```
